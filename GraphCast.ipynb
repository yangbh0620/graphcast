```python
# @title Plotting functions

def select(
    data: xarray.Dataset,
    variable: str,
    level: Optional[int] = None,
    max_steps: Optional[int] = None
    ) -> xarray.Dataset:
  data = data[variable]
  if "batch" in data.dims:
    data = data.isel(batch=0)
  if max_steps is not None and "time" in data.sizes and max_steps < data.sizes["time"]:
    data = data.isel(time=range(0, max_steps))
  if level is not None and "level" in data.coords:
    data = data.sel(level=level)
  return data

def scale(
    data: xarray.Dataset,
    center: Optional[float] = None,
    robust: bool = False,
    ) -> tuple[xarray.Dataset, matplotlib.colors.Normalize, str]:
  vmin = np.nanpercentile(data, (2 if robust else 0))
  vmax = np.nanpercentile(data, (98 if robust else 100))
  if center is not None:
    diff = max(vmax - center, center - vmin)
    vmin = center - diff
    vmax = center + diff
  return (data, matplotlib.colors.Normalize(vmin, vmax),
          ("RdBu_r" if center is not None else "viridis"))

def plot_data(
    data: dict[str, xarray.Dataset],
    fig_title: str,
    plot_size: float = 5,
    robust: bool = False,
    cols: int = 4
    ) -> tuple[xarray.Dataset, matplotlib.colors.Normalize, str]:

  first_data = next(iter(data.values()))[0]
  max_steps = first_data.sizes.get("time", 1)
  assert all(max_steps == d.sizes.get("time", 1) for d, _, _ in data.values())

  cols = min(cols, len(data))
  rows = math.ceil(len(data) / cols)
  figure = plt.figure(figsize=(plot_size * 2 * cols,
                               plot_size * rows))
  figure.suptitle(fig_title, fontsize=16)
  figure.subplots_adjust(wspace=0, hspace=0)
  figure.tight_layout()

  images = []
  for i, (title, (plot_data, norm, cmap)) in enumerate(data.items()):
    ax = figure.add_subplot(rows, cols, i+1)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(title)
    im = ax.imshow(
        plot_data.isel(time=0, missing_dims="ignore"), norm=norm,
        origin="lower", cmap=cmap)
    plt.colorbar(
        mappable=im,
        ax=ax,
        orientation="vertical",
        pad=0.02,
        aspect=16,
        shrink=0.75,
        cmap=cmap,
        extend=("both" if robust else "neither"))
    images.append(im)

  def update(frame):
    if "time" in first_data.dims:
      td = datetime.timedelta(microseconds=first_data["time"][frame].item() / 1000)
      figure.suptitle(f"{fig_title}, {td}", fontsize=16)
    else:
      figure.suptitle(fig_title, fontsize=16)
    for im, (plot_data, norm, cmap) in zip(images, data.values()):
      im.set_data(plot_data.isel(time=frame, missing_dims="ignore"))

  ani = animation.FuncAnimation(
      fig=figure, func=update, frames=max_steps, interval=250)
  plt.close(figure.number)
  return HTML(ani.to_jshtml())
```

# Load the Data and initialize the model

## Load the model params

Choose one of the two ways of getting model params:
- **random**: You'll get random predictions, but you can change the model architecture, which may run faster or fit on your device.
- **checkpoint**: You'll get sensible predictions, but are limited to the model architecture that it was trained with, which may not fit on your device. In particular generating gradients uses a lot of memory, so you'll need at least 25GB of ram (TPUv4 or A100).

Checkpoints vary across a few axes:
- The mesh size specifies the internal graph representation of the earth. Smaller meshes will run faster but will have worse outputs. The mesh size does not affect the number of parameters of the model.
- The resolution and number of pressure levels must match the data. Lower resolution and fewer levels will run a bit faster. Data resolution only affects the encoder/decoder.
- All our models predict precipitation. However, ERA5 includes precipitation, while HRES does not. Our models marked as "ERA5" take precipitation as input and expect ERA5 data as input, while model marked "ERA5-HRES" do not take precipitation as input and are specifically trained to take HRES-fc0 as input (see the data section below).

We provide three pre-trained models.
1. `GraphCast`, the high-resolution model used in the GraphCast paper (0.25 degree resolution, 37 pressure levels), trained on ERA5 data from 1979 to 2017,

2. `GraphCast_small`, a smaller, low-resolution version of GraphCast (1 degree resolution, 13 pressure levels, and a smaller mesh), trained on ERA5 data from 1979 to 2015, useful to run a model with lower memory and compute constraints,

3. `GraphCast_operational`, a high-resolution model (0.25 degree resolution, 13 pressure levels) pre-trained on ERA5 data from 1979 to 2017 and fine-tuned on HRES data from 2016 to 2021. This model can be initialized from HRES data (does not require precipitation inputs).

## Load the example data

Several example datasets are available, varying across a few axes:
- **Source**: fake, era5, hres
- **Resolution**: 0.25deg, 1deg, 6deg
- **Levels**: 13, 37
- **Steps**: How many timesteps are included

Not all combinations are available.
- Higher resolution is only available for fewer steps due to the memory requirements of loading them.
- HRES is only available in 0.25 deg, with 13 pressure levels.

The data resolution must match the model that is loaded.

Some transformations were done from the base datasets:
- We accumulated precipitation over 6 hours instead of the default 1 hour.
- For HRES data, each time step corresponds to the HRES forecast at leadtime 0, essentially providing an "initialisation" from HRES. See HRES-fc0 in the GraphCast paper for further description. Note that a 6h accumulation of precipitation is not available from HRES, so our model taking HRES inputs does not depend on precipitation. However, because our models predict precipitation, we include the ERA5 precipitation in the example data so it can serve as an illustrative example of ground truth.
- We include ERA5 `toa_incident_solar_radiation` in the data. Our model uses the radiation at -6h, 0h and +6h as a forcing term for each 1-step prediction. If the radiation is missing from the data (e.g. in an operational setting), it will be computed using a custom implementation that produces values similar to those in ERA5.

```python
# @title Build jitted functions, and possibly initialize random weights

def construct_wrapped_graphcast(
    model_config: graphcast.ModelConfig,
    task_config: graphcast.TaskConfig):
  """Constructs and wraps the GraphCast Predictor."""
  # Deeper one-step predictor.
  predictor = graphcast.GraphCast(model_config, task_config)

  # Modify inputs/outputs to `graphcast.GraphCast` to handle conversion to
  # from/to float32 to/from BFloat16.
  predictor = casting.Bfloat16Cast(predictor)

  # Modify inputs/outputs to `casting.Bfloat16Cast` so the casting to/from
  # BFloat16 happens after applying normalization to the inputs/targets.
  predictor = normalization.InputsAndResiduals(
      predictor,
      diffs_stddev_by_level=diffs_stddev_by_level,
      mean_by_level=mean_by_level,
      stddev_by_level=stddev_by_level)

  # Wraps everything so the one-step model can produce trajectories.
  predictor = autoregressive.Predictor(predictor, gradient_checkpointing=True)
  return predictor


@hk.transform_with_state
def run_forward(model_config, task_config, inputs, targets_template, forcings):
  predictor = construct_wrapped_graphcast(model_config, task_config)
  return predictor(inputs, targets_template=targets_template, forcings=forcings)


@hk.transform_with_state
def loss_fn(model_config, task_config, inputs, targets, forcings):
  predictor = construct_wrapped_graphcast(model_config, task_config)
  loss, diagnostics = predictor.loss(inputs, targets, forcings)
  return xarray_tree.map_structure(
      lambda x: xarray_jax.unwrap_data(x.mean(), require_jax=True),
      (loss, diagnostics))

def grads_fn(params, state, model_config, task_config, inputs, targets, forcings):
  def _aux(params, state, i, t, f):
    (loss, diagnostics), next_state = loss_fn.apply(
        params, state, jax.random.PRNGKey(0), model_config, task_config,
        i, t, f)
    return loss, (diagnostics, next_state)
  (loss, (diagnostics, next_state)), grads = jax.value_and_grad(
      _aux, has_aux=True)(params, state, inputs, targets, forcings)
  return loss, diagnostics, next_state, grads

# Jax doesn't seem to like passing configs as args through the jit. Passing it
# in via partial (instead of capture by closure) forces jax to invalidate the
# jit cache if you change configs.
def with_configs(fn):
  return functools.partial(
      fn, model_config=model_config, task_config=task_config)

# Always pass params and state, so the usage below are simpler
def with_params(fn):
  return functools.partial(fn, params=params, state=state)

# Our models aren't stateful, so the state is always empty, so just return the
# predictions. This is requiredy by our rollout code, and generally simpler.
def drop_state(fn):
  return lambda **kw: fn(**kw)[0]

init_jitted = jax.jit(with_configs(run_forward.init))

if params is None:
  params, state = init_jitted(
      rng=jax.random.PRNGKey(0),
      inputs=train_inputs,
      targets_template=train_targets,
      forcings=train_forcings)

loss_fn_jitted = drop_state(with_params(jax.jit(with_configs(loss_fn.apply))))
grads_fn_jitted = with_params(jax.jit(with_configs(grads_fn)))
run_forward_jitted = drop_state(with_params(jax.jit(with_configs(
    run_forward.apply))))
```

This cell attempts to write the custom loss functions into
`graphcast/extreme_losses.py` so they can be imported as a module.

However, since the current environment does not contain an actual Python
package named `graphcast`, the module import will not work.  
For reproducibility and to ensure the notebook runs end-to-end, the loss
functions are instead defined directly inside the notebook in a later cell.

```python
%%writefile graphcast/extreme_losses.py
"""
Extreme-aware latitude-weighted loss functions for GraphCast.

This module provides:
    - A clean implementation of the original latitude-weighted MSE.
    - An extended version that applies additional weights to extreme targets.

Adding this file constitutes a replacement/extension of one modular block in
the original GraphCast/GenCast system (loss function module).
"""

import jax.numpy as jnp


def _broadcast_lat_weights(lat, error):
    """
    Broadcast 1D latitude weights to match the shape of the error array.

    Args:
        lat: 1D array of latitudes [lat]
        error: array of shape (..., lat, lon)

    Returns:
        An array of shape (..., lat, lon) containing the broadcasted
        cosine(latitude) weights.
    """
    # Latitude weights (cosine weighting)
    lat_w = jnp.cos(jnp.deg2rad(lat))               # [lat]
    lat_w = lat_w / (jnp.mean(lat_w) + 1e-12)       # Normalize

    # Broadcast to match the error shape: (..., lat, lon)
    w = lat_w
    while w.ndim < error.ndim - 1:
        w = w[jnp.newaxis, ...]
    w = w[..., :, jnp.newaxis]                      # Add lon dimension

    return w


def lat_weighted_mse(prediction, target, latitude):
    """
    Original latitude-weighted mean squared error (MSE).

    This is the baseline objective used by GraphCast. The purpose of this
    implementation is to provide a clean reference function for comparison.

    Args:
        prediction: Predicted fields, shape (..., lat, lon)
        target: Ground-truth fields, same shape
        latitude: 1D latitude array [lat]

    Returns:
        Scalar latitude-weighted MSE.
    """
    error = prediction - target
    weight = _broadcast_lat_weights(latitude, error)

    loss_num = jnp.sum(weight * (error ** 2))
    loss_den = jnp.sum(weight) + 1e-12
    return loss_num / loss_den


def lat_weighted_mse_extreme(
    prediction,
    target,
    latitude,
    *,
    extreme_threshold=2.0,
    extreme_multiplier=5.0,
):
    """
    Extreme-aware latitude-weighted MSE.

    This loss enhances the contribution of rare, high-impact events.
    Points where |target| >= extreme_threshold are multiplied by an
    additional factor (extreme_multiplier).

    Args:
        prediction: Predicted fields, shape (..., lat, lon)
        target: Ground-truth fields, same shape
        latitude: 1D latitude array [lat]
        extreme_threshold: |y| >= threshold is considered 'extreme'
        extreme_multiplier: Weight multiplier for extreme events (alpha > 1)

    Returns:
        Scalar extreme-aware latitude-weighted MSE.
    """
    error = prediction - target

    # Baseline latitude weights
    base_weight = _broadcast_lat_weights(latitude, error)

    # Extreme event mask
    extreme_mask = (jnp.abs(target) >= extreme_threshold).astype(jnp.float32)

    # Final weights: w = w_lat * [1 + (alpha - 1) * mask]
    weight = base_weight * (
        1.0 + (extreme_multiplier - 1.0) * extreme_mask
    )

    loss_num = jnp.sum(weight * (error ** 2))
    loss_den = jnp.sum(weight) + 1e-12
    return loss_num / loss_den
```

### Cell: Define baseline and extreme-aware loss functions

This cell implements the custom loss functions used in my experiments:

- `_broadcast_lat_weights(latitude, error)`  
  builds cosine(latitude) weights and broadcasts them to match the
  spatial shape of the prediction error tensor.

- `lat_weighted_mse(prediction, target, latitude)`  
  reproduces the original **latitude-weighted mean squared error (MSE)**
  used as the baseline objective in GraphCast, where each grid point is
  weighted by the cosine of its latitude.

- `lat_weighted_mse_extreme(prediction, target, latitude, ...)`  
  defines a new **extreme-aware latitude-weighted MSE**. In addition to
  the latitude weights, grid points with large absolute targets
  (|target| ≥ 2σ by default) are up-weighted by a factor
  `extreme_multiplier` (e.g., 5.0). This function is my replacement of
  the original loss module and is later used to compare baseline vs.
  extreme-focused behavior.


```python
import jax.numpy as jnp

def _broadcast_lat_weights(latitude, error):
    """
    Broadcast 1D latitude weights to match the shape of the error array.

    Args:
        latitude: 1D array of latitudes in degrees, shape [lat].
        error: Array of prediction errors, shape (..., lat, lon).

    Returns:
        Weight array of shape (..., lat, lon) based on cosine(latitude),
        normalized by its global mean.
    """
    # Cosine latitude weights
    lat_w = jnp.cos(jnp.deg2rad(latitude))            # [lat]
    lat_w = lat_w / (jnp.mean(lat_w) + 1e-12)         # Normalize

    # Broadcast to match error shape: (..., lat, lon)
    w = lat_w
    while w.ndim < error.ndim - 1:
        w = w[jnp.newaxis, ...]
    w = w[..., :, jnp.newaxis]                        # Add lon dimension
    return w


def lat_weighted_mse(prediction, target, latitude):
    """
    Original latitude-weighted mean squared error (MSE).

    This function reproduces the baseline objective used in GraphCast,
    where each grid point is weighted by the cosine of its latitude.

    Args:
        prediction: Predicted fields, shape (..., lat, lon).
        target: Ground-truth fields, same shape as prediction.
        latitude: 1D latitude array in degrees, shape [lat].

    Returns:
        A scalar latitude-weighted MSE.
    """
    error = prediction - target
    weight = _broadcast_lat_weights(latitude, error)

    loss_num = jnp.sum(weight * (error ** 2))
    loss_den = jnp.sum(weight) + 1e-12
    return loss_num / loss_den


def lat_weighted_mse_extreme(
    prediction,
    target,
    latitude,
    *,
    extreme_threshold=2.0,
    extreme_multiplier=5.0,
):
    """
    Extreme-aware latitude-weighted MSE.

    Compared to the original latitude-weighted MSE, this loss increases
    the contribution of rare, high-impact events. Points where
    |target| >= extreme_threshold are up-weighted by 'extreme_multiplier'.

    Args:
        prediction: Predicted fields, shape (..., lat, lon).
        target: Ground-truth fields, same shape as prediction.
        latitude: 1D latitude array in degrees, shape [lat].
        extreme_threshold: Absolute target value above which a point is
            considered 'extreme' (e.g., 2.0 for 2 standard deviations).
        extreme_multiplier: Multiplicative factor for extreme points
            (alpha > 1 makes extremes more important).

    Returns:
        A scalar extreme-aware latitude-weighted MSE.
    """
    error = prediction - target

    # Baseline cosine(latitude) weights
    base_weight = _broadcast_lat_weights(latitude, error)

    # Binary mask selecting extreme targets
    extreme_mask = (jnp.abs(target) >= extreme_threshold).astype(jnp.float32)

    # Final weights: w = w_lat * [1 + (alpha - 1) * mask]
    weight = base_weight * (
        1.0 + (extreme_multiplier - 1.0) * extreme_mask
    )

    loss_num = jnp.sum(weight * (error ** 2))
    loss_den = jnp.sum(weight) + 1e-12
    return loss_num / loss_den
```

prediction error using root-mean-square error (RMSE) in two ways:

- **Overall RMSE** computed over all grid points, and  
- **Extreme-only RMSE** computed only on the subset of grid points where
  |target| ≥ 2σ (the same threshold used in the extreme-aware loss).

and they show how the model performs specifically on extreme events
compared to the full domain.

```python
threshold = 2.0

error = pred_var - tgt_var

# Overall RMSE
rmse_all = jnp.sqrt(jnp.mean(error ** 2))

# RMSE on extreme subset (|y| >= threshold)
mask_extreme = jnp.abs(tgt_var) >= threshold
rmse_extreme = jnp.sqrt(jnp.mean((error ** 2)[mask_extreme]))

print("Variable:", var_name)
print("Overall RMSE              :", float(rmse_all))
print(f"RMSE on |y| >= {threshold}σ targets:", float(rmse_extreme))
```

### Cell: Generate comparison table for multiple variables

This cell evaluates the baseline and extreme-aware loss functions across
three different variables (`2m_temperature`, `10m wind`, and
`mean sea level pressure`). For each variable, the following metrics are
computed:

- **Baseline MSE**: the original latitude-weighted MSE used by GraphCast  
- **Extreme-loss MSE**: the newly proposed loss that up-weights |target| ≥ 2σ  
- **Global RMSE**: root-mean-square error over all grid points  
- **Extreme-only RMSE**: RMSE computed only on the extreme subset  
  (|target| ≥ 2σ)

All results are collected into a Pandas DataFrame (`df_results`), which
serves as the main comparison table for the assignment. This table is
later used in the report to discuss the behavior of the new loss function.


```python
import pandas as pd

variables = [
    "2m_temperature",
    "10m_u_component_of_wind",
    "mean_sea_level_pressure",
]

rows = []
for v in variables:
    pred_v = jnp.asarray(predictions[v].values)
    tgt_v  = jnp.asarray(eval_targets[v].values)

    base_loss = lat_weighted_mse(pred_v, tgt_v, lat)
    ext_loss  = lat_weighted_mse_extreme(pred_v, tgt_v, lat,
                                         extreme_threshold=2.0,
                                         extreme_multiplier=5.0)

    error = pred_v - tgt_v
    rmse_all = jnp.sqrt(jnp.mean(error ** 2))
    mask_extreme = jnp.abs(tgt_v) >= 2.0
    rmse_extreme = jnp.sqrt(jnp.mean((error ** 2)[mask_extreme]))

    rows.append({
        "Variable": v,
        "Baseline MSE": float(base_loss),
        "Extreme-loss MSE": float(ext_loss),
        "Global RMSE": float(rmse_all),
        "Extreme-only RMSE": float(rmse_extreme),
    })

df_results = pd.DataFrame(rows)
df_results
```
